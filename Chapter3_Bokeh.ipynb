# DATA VISUALIZATION WITH BOKEH

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, output_file, show, output_notebook
from bokeh.io import output_file, show
from bokeh.plotting import ColumnDataSource
output_notebook()
```

```python
x =np.array([1,2,3,4,5,6])
y = np.array([5,6,7,8,9,10])
p = figure(plot_width=300, plot_height=200, tools="pan,hover")
p.line(x,y)
show(p)
```

```python
x_red = np.array([1,2,3,4,5])
y_red = np.array([5,6,7,8,9])

x_blue = np.array([10,11,12,13])
y_blue = np.array([14,15,16,17])

#Creating the categorical scatter plot 

plot = figure(plot_width=300, plot_height=200)

plot.circle(x_red, y_red, size = 9, color = 'red', alpha = 0.8)
plot.circle(x_blue, y_blue, size = 9, color = 'blue', alpha = 0.8)
show(plot)
```

```python
df = pd.read_csv('all_stocks_5yr.csv')
```

```python
from datetime import date
from datetime import datetime
```

```python
df.head()
```

```python
df['date']=pd.to_datetime(df['date'], infer_datetime_format=True)
```

```python
df_apple = df[df['Name'] == 'AAL']
x=df_apple['date']
y=df_apple['high']
x.head()
#y.head()
```

```python
import math
```

```python
plot = figure(plot_width=400, plot_height=200,
                x_axis_type = 'datetime', 
              x_axis_label = 'date', 
              y_axis_label = 'High Prices')
plot.xaxis.major_label_orientation = math.pi/4

plot.line(x,y)
show(plot)
```

```python
plot = figure(plot_width=400, plot_height=200)
#plot.line(x.index, y = df_apple['high'])

plot.circle(x=df_apple['close'],y=df_apple['open'])
show(plot)
```

```python
plot = figure(plot_width=400, plot_height=200)
#plot.line(x.index, y = df_apple['high'])

plot.diamond(x=df_apple['close'],y=df_apple['open'],color='red',
            size=5,alpha=0.7)
show(plot)
```

```python
data = ColumnDataSource(df_apple)
plot = figure(plot_width=400, plot_height=200,
    x_axis_type = 'datetime', x_axis_label = 'date', y_axis_label = 'High Prices')

#plot.line(x = 'low', y = 'high', source = data, color = 'red')

plot.circle(x = 'low', y = 'high', source = data, fill_color = 'white', size = 3)
show(plot)
```
